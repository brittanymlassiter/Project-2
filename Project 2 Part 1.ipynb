## Preprocesing data

### Load Data

```python
df = pd.read_csv('Data/title-akas-us-only.csv', low_memory=False)
```

```python
df.head()
```

```python
basics = pd.read_csv('Data/title.basics.tsv.gz', sep='\t', low_memory=False)
```

```python
basics.head()
```

```python
# Filter the basics table down to only include the US by using the filter akas dataframe
filter_us_titles = basics['tconst'].isin(df['titleId'])
basics = basics[filter_us_titles]
basics
```

```python
ratings=pd.read_csv('Data/title.ratings.tsv.gz', sep='\t', low_memory=False)
ratings.head()
```

### Replacing/Dropping Null Values

```python
# Replace Missing values
basics = basics.replace({'\\N':np.nan})
```

```python
#Drop the Nulls in Rnutimeminutes and genres
basics = basics.dropna(subset=["runtimeMinutes", "genres"])

# Confirming no more null sale prices
basics['runtimeMinutes'].isna().sum()
basics['genres'].isna().sum()
```

### Filtering out the Movies only

```python
# Filtering to only have title type ==Movie
basics.loc[basics['titleType'] == 'movie']

#Save as a new Variable
filter_movie = basics["titleType"] == 'movie'
basics= basics[filter_movie]
```

```python
# Apply the filter to create a dataframe of selected values using square brackets
basics_filtered = basics[filter_movie]
basics_filtered
```

### Convert startYear to a float dtype/Filter to keep movies 2000-2022

```python
# Convert a Pandas DataFrame Column From String to Float
basics['startYear'] = basics['startYear'].astype(float)
basics['startYear'].dtypes
```

```python
# Filtering to only have years 2000-2022
basics = basics[(basics['startYear']>=2000) & (basics['startYear']<=2022)]
```

### Eliminate movies that include "Documentary" in genre

```python
# Eliminate movies that include "Documentary" 
filter_documentaries = basics['genres'].str.contains('Documentary')
```

```python
# Exclude movies in the documentary category.
basics = basics[~filter_documentaries]
```

```python
# load data and keep only movies
filter_basics = ratings['tconst'].isin(basics['tconst'])
ratings = ratings[filter_basics]
ratings
```
